```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data= pd.DataFrame({
'MONTH': [0,1,2,3,4,5,6,7,8],
'Rain': [10,30,20,50,60,30,40,10,30]
})
print(data)
```

```python
data.plot(y='Rain', x='MONTH',color="#ADD8E6",linestyle='--',marker="*",markerfacecolor="red",markersize=12)
plt.ylabel=('Rain')
plt.xlabel=('Month')
plt.ylabel=('Rain')
plt.title=('Rain in every month')
plt.show()
```
